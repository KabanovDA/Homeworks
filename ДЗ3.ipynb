```python
import numpy as np
import matplotlib.pyplot as plt


data = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')

fig = plt.figure()
plt.figure(figsize=(10,8))
plt.xticks(range(1900,1921))
fig.patch.set_facecolor('xkcd:white')

Y = data[:,0:1]
A = data[:,1:2]
B = data[:,2:3]
C = data[:,3:4]


plt.plot(Y, A, label='hare', c='r', linestyle='-', linewidth=2)

plt.plot(Y, B, label='lynx', c='g', linestyle='-', linewidth=2)

plt.plot(Y, C, label='carrot', c='b', linestyle='-', linewidth=2)

plt.xlabel('Year', fontsize=10)
plt.ylabel('Count')
plt.xticks(fontsize=9)
plt.legend()
plt.show()
```

```python
fig = plt.figure()

plt.figure(figsize=(10,8))
plt.xticks(range(1900,1921))
fig.patch.set_facecolor('xkcd:white')
plt.plot(Y, A, label='hare', c='r', linestyle='-', linewidth=2)
plt.plot(Y, B, label='lynx', c='g', linestyle='-', linewidth=2)
plt.plot(Y, C, label='carrot', c='b', linestyle='-', linewidth=2)
for i in range(3):
  y = data[:,(i+1):(i+2)]
  xpos = np.argmax(y)
  ymax = y[xpos]
  xmax = Y[xpos]
  plt.scatter(xmax, ymax, color='black', s=100, marker='o')
  plt.annotate(f'{int(xmax)},{int(ymax)}', (xmax, ymax))

plt.xlabel('Year', fontsize=10)
plt.ylabel('Count')
plt.xticks(fontsize=9)
plt.legend()
plt.show()
```

```python
from matplotlib import legend
import matplotlib.pyplot as plt
width = 0.9

fig, ax = plt.subplots(1,3, sharex=True, sharey=True, figsize = (20,8))


yy = [str(int(y)) for y in Y] 

ax[0].axhline(y=A.mean(),c="black",linewidth=0.5,zorder=0)
ax[0].axhline(y=A.std(),c="grey",linewidth=0.5,zorder=0)
ax[0].bar(yy, A.reshape(-1), width, label='hare', color='red')

ax[1].axhline(y=B.mean(),c="black",linewidth=0.5,zorder=0)
ax[1].axhline(y=B.std(),c="grey",linewidth=0.5,zorder=0)
ax[1].bar(yy, B.reshape(-1), width, label='lynx', color='green')

ax[2].axhline(y=C.mean(),c="black",linewidth=0.5,zorder=0)
ax[2].axhline(y=C.std(),c="grey",linewidth=0.5,zorder=0)
ax[2].bar(yy, C.reshape(-1), width, label='carrot', color='blue')

ax[0].set_xlabel('Year')
ax[0].set_ylabel('Count')
ax[0].set_ylabel('population')
ax[0].legend()

plt.sca(ax[0])
plt.xticks(rotation=90)

ax[1].set_xlabel('Year')
ax[1].set_ylabel('Count')
ax[1].set_ylabel('population hare')
ax[1].legend()
plt.sca(ax[1])
plt.xticks(rotation=90)

ax[2].set_xlabel('Year')
ax[2].set_ylabel('Count lynx')
ax[2].set_ylabel('population carrot')
ax[2].legend()
plt.sca(ax[2])
plt.xticks(rotation=90)


plt.show()
```
